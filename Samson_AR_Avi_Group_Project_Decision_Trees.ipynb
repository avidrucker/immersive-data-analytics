```python
import pandas as pd
```

```python
location = "datasets/OULAD.xlsx"
df = pd.read_excel(location) #overwrites the info from the df variable in the examples above
print(df.shape)
df.head()
```

```python
df['id_mod_pre_stu_ass'] = df['id_course'] + "-" + str(df['id_student.1'][0]) + "-" + str(df["id_assessment"][0])
```

```python
df1 = df.copy()
df1.head()
```

```python
df2 = df1.copy()
df2 = df2.drop(['id_student','Unnamed: 1','id_student.1','id_course','id_assessment'],axis=1)
df2.head()
```

```python
df2['id_mod_pre_stu_ass'].value_counts()[:5]
```

```python
df3 = df2.copy()
df3.sort_values('date_submitted', inplace=True)
df3 = df3.groupby('id_mod_pre_stu_ass', as_index=False).first()
#df3['id_mod_pre_stu_ass'].value_counts()
df3.head()
```

```python
# remove assessement id from id_mod_pre_stu_ass
df4 = df3.copy()
df4['id_assessment'] = df4['id_mod_pre_stu_ass'].str.split("-")
df4['id_mod_pre_stu'] = df4['id_mod_pre_stu_ass'].str.split("-")
df4['id_assessment'] = df4['id_assessment'].apply(lambda x: x[3])
df4['id_mod_pre_stu'] = df4['id_mod_pre_stu'].apply(lambda x: x[0] + '-' + x[1] + "-" + str(x[2]))
df4 = df4.drop("id_mod_pre_stu_ass",axis=1)
#df4 = df4.rename(columns={'id_mod_pre_stu_ass': 'id_mod_pre_stu'})

df4.id_mod_pre_stu = df4.id_mod_pre_stu.astype(str)
#df4.id_mod_pre_stu = df4.id_mod_pre_stu.str.encode('utf-8')

df4.head()
```

```python
print(df4.shape)
df4['id_mod_pre_stu'].value_counts()[:5]
```
